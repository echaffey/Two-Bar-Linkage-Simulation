```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('data2.xlsx')
```

### Crank

```python
plt.figure(figsize=(8,8));
fig,ax = plt.subplots(3,1,sharex=True);
ax[0].plot(df.time,df.cr_pos,'b');
ax[1].plot(df.time,df.cr_vel,'g');
ax[2].plot(df.time,np.zeros_like(df.time),'r');
```

### Connecting Rod

```python
plt.figure(figsize=(8,8))
fig,ax = plt.subplots(3,1,sharex=True)
ax[0].plot(df.time,df.con_pos,'b')
ax[1].plot(df.time,df.con_vel,'g')
ax[2].plot(df.time,df.con_acc,'r');
```

### Piston

```python
plt.figure(figsize=(8,8))
fig,ax = plt.subplots(3,1,sharex=True)
ax[0].plot(df.time,df.p_pos,'b')
ax[1].plot(df.time,df.p_vel,'g')
ax[2].plot(df.time,df.p_acc,'r');
```
