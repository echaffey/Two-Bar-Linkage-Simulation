```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('data.xlsx')
df.head()
```

```python
plt.figure(figsize=(8,6))
plt.plot(df.time, df.position, 'b-o', ms=4)
plt.title('Piston Position',fontsize=14)
plt.ylabel('Displacement (in)')
plt.xlabel('Time (s)')
plt.grid()
plt.savefig('displacement.png',dpi=300,bbox_inches='tight')
```

```python
plt.figure(figsize=(8,6))
plt.plot(df.time, df.velocity, 'g-o', ms=4)
plt.title('Piston Velocity',fontsize=14)
plt.ylabel('Speed (in/s)')
plt.xlabel('Time (s)')
plt.grid()
plt.savefig('velocity.png',dpi=300,bbox_inches='tight')
```

```python
plt.figure(figsize=(8,6))
plt.plot(df.time[1:], df.acceleration[1:], 'r-o', ms=4)
plt.title('Piston Acceleration',fontsize=14)
plt.ylabel('Acceleration $(in/s^2)$')
plt.xlabel('Time (s)')
plt.grid()
plt.savefig('acceleration.png',dpi=300,bbox_inches='tight')
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(3, figsize=(8,15))
ax1.plot(df.time, df.position, 'b-o', ms=4)
ax2.plot(df.time, df.velocity, 'g-o', ms=4)
ax3.plot(df.time[1:], df.acceleration[1:], 'r-o', ms=4)
ax1.grid()
ax2.grid()
ax3.grid()
```
